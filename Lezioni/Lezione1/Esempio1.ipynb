```python
class DocumentAutomationError(Exception):
    """Classe base per le eccezioni del nostro software."""
    pass

class EmptyDocumentError(DocumentAutomationError):
    """Lanciata quando un file OCR non produce testo."""
    pass

class APIConnectionError(DocumentAutomationError):
    """Lanciata in caso di fallimento persistente delle API."""
    pass
```

```python
def extract_data(text):
    if not text.strip():
        raise EmptyDocumentError("Il testo estratto è vuoto, verificare qualità immagine.")
    return {"content": text.strip().upper()}

def run_pipeline(docs):
    for i, doc in enumerate(docs):
        print(f"Processando documento {i}...")
        try:
            result = extract_data(doc)
            print(f"Successo: {result}")
        except EmptyDocumentError as e:
            print(f"Errore specifico: {e}")
        except Exception as e:
            print(f"Errore imprevisto: {e}")
lista_testi = ["Fattura 01", "  ", "Fattura 02"]
run_pipeline(lista_testi)
```
